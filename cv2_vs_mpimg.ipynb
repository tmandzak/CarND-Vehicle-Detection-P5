### mpimg JPEG

```python
img1 = mpimg.imread("experiments/cutout1.jpg")
img1 = img1.astype(np.float32)/255
print(np.min(img1), np.max(img1))
img1 = cv2.cvtColor(img1, cv2.COLOR_RGB2YUV)
print(np.min(img1), np.max(img1))
```
